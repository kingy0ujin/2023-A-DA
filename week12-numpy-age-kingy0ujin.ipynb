# numpy.hist()

```python
import matplotlib.pyplot as plt
import numpy as np

data = [1,1,2,3,4,5,7,8,9,10]

plt.hist(data)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [1,1,2,3,4,5,7,8,9,10]

plt.hist(data, bins=10, rwidth=.9) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = [1,1,2,3,4,5,7,8,9,10]

plt.hist(data, bins=3, rwidth=.9) # bins -> 구간이 몇 구간인지, 결과는 구간에 포함된 수로 막대의 높이가 결정됨
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)

mask1 = (abs(x) > 50)
mask2 = (abs(y) > 50)
x = x[mask1 + mask2]
y = y[mask1 + mask2]

plt.scatter(x, y, c=x, cmap='jet', alpha = 0.7 )
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)

mask1 = (abs(x) > 50)
mask2 = (abs(y) > 50)
x = x[mask1 * mask2]
y = y[mask1 * mask2]

plt.scatter(x, y, c=x, cmap='jet', alpha = 0.7 )
plt.colorbar()
plt.show()
```
