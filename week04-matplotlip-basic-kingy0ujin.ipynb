```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N , endpoint=True) #10을 포함해서 8개 수로 등분, 전체 7등분
x2 = np.linspace(0, 10, N , endpoint=False) #10을 빼고 8개 수로 등분, 전체 8등분
plt.plot(x1, y + 0.5, 'o', label='endpoint=True')
plt.plot(x2, y, 'o', label='endpoint=False')

plt.legend()
plt.ylim([-0.5, 1])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = 8
y = np.zeros(N)

x1 = np.linspace(0, 10, N , endpoint=True)
x2 = np.linspace(0, 10, N , endpoint=False)
plt.plot(x1, y + 0.5, '-o', label='endpoint=True')
plt.plot(x2, y, 'o:', label='endpoint=False')

plt.legend()
plt.ylim([-0.5, 1]) # y축 범위 지정
plt.show()
```
